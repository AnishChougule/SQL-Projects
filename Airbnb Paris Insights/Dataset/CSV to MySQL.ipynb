```python
import pandas as pd
import numpy as np
import glob
from sqlalchemy import create_engine
```

```python
df_list = pd.read_csv("listings.csv")
df_rev = pd.read_csv("reviews.csv")
```

```python
df_list['price'] = df_list['price'].str.replace('$', '').str.replace(',', '')
df_list['bathrooms_text'] = df_list['bathrooms_text'].str.replace('[^\d.]', '', regex=True)
df_list = df_list.rename(columns={'id': 'listing_id', 'bathrooms_text': 'bathrooms'})
df_list = df_list.replace('', None)
```

```python
# Replacing infinite values with None in all DataFrames
dfs = [df_list, df_rev]
for df in dfs:
    df.replace([np.inf, -np.inf], None, inplace=True)
```

```python
host = 'localhost'
port = 3306
user = 'root'
password = 'root'
database = 'airbnb_paris'
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")

# Convert the DataFrame to an SQL table
df_list.to_sql('listings', engine, if_exists='replace', index=False)
df_rev.to_sql('reviews', engine, if_exists='replace', index=False)

# Print a success message
print('DataFrame converted to SQL successfully!')
```
